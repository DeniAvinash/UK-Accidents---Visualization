```python
data['Date'] = data['Date'].map(pandas.to_datetime)
```

```python
def get_dom(dt):
    return dt.day

data['dom'] = data['Date'].map(get_dom)
```

```python
data.tail()
```

```python
def get_weekday(dt):
    return dt.weekday()

data['weekday'] = data['Date'].map(get_weekday)
```
